### BIVARIATE ANALYSIS ---

Bivariate analysis is the statistical analysis of two variables to determine the relationship between them. It helps in understanding correlations, trends, and patterns in data.
Means one variable dependency in another variable.

```python
df_new.columns
```

```python
# Price Dependency on neighbourhood_group

sns.barplot(data = df_new, x='neighbourhood_group',y='price_per_bed', hue='room_type')
```

```python
plt.figure(figsize=(8,5))
plt.title('Review Distribution')
sns.scatterplot(data = df_new, x='number_of_reviews',y='price',hue='neighbourhood_group')
plt.show()
```

```python
df_new.dtypes
```

```python
plt.figure(figsize=(8,6))
sns.pairplot(data = df_new, vars=['price','minimum_nights','number_of_reviews','availability_365'], hue='room_type')
```

```python
plt.figure(figsize=(10,7))
plt.title('GEOGRAPHICAL DISTRIBUTION OF AIRBNB LISTING')
sns.scatterplot(data=df_new, x='longitude', y='latitude',hue='room_type')
plt.show()
```

```python
df_new.dtypes
```

```python
# HEAT MAP -- CORRELATION OF ONE VARIABLE WITH OTHERS FOR NUMERICAL COLUMN
# CREATING CORRELATION -------(-1 TO 1)

    # +1 → Perfect positive correlation (both variables increase together).
    # 0 → No correlation (no relationship between variables).
    # -1 → Perfect negative correlation (one increases while the other decreases). 

corr_matrix = df_new[['latitude','longitude','price','minimum_nights','number_of_reviews','reviews_per_month', 
        'availability_365', 'beds']].corr()
corr_matrix


plt.figure(figsize=(9,6))
sns.heatmap(data = corr_matrix, annot=True)

# Here, price vs beds is highlighted with red color means it's a perfect positive correlation as 
    # when price increases no. of beds also increases.
```
